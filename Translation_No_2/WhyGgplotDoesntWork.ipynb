```python
a = np.arange(1,100,1)
b = (a + np.random.randn(99))**2

df1 = pd.DataFrame({"col1":a, "col2":b})
df1
```

```python
# das hier sollte laut tutorial schon funktionieren. Chart bleibt aber leer
# documentation: http://ggplot.yhathq.com/how-it-works.html
p = ggplot(aes(x = "col1", y = "col2"), data = df1) 
p
```

```python
# erst durch hinzufÃ¼gen solcher befehle erscheint die volle grafik
p + geom_point() + geom_line() 
```
